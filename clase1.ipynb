```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("lena.png")
#img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

plt.imshow(img)

plt.show()
```

```python
blue, green, red = cv2.split(img)

plt.imshow(red, cmap="gray")
plt.show()
```

```python
plt.imshow(green, cmap="hot")
plt.show()
```

```python
plt.imshow(blue, cmap="winter")
plt.show()
```

```python
hist_red = cv2.calcHist([img],[2], None, [256], [0,256])
hist_green = cv2.calcHist([img],[1], None, [256], [0,256])
hist_blue = cv2.calcHist([img],[0], None, [256], [0,256])

plt.plot(hist_red, color="red", label="canal Rojo")
plt.plot(hist_green, color="green", label="canal Verde")
plt.plot(hist_blue, color="blue", label="canal Azul")

plt.title("Histograma de Lena")
plt.xlabel("Niveles de intensidad")
plt.ylabel("Frecuencia")

plt.xlim([0,256])
plt.legend()
plt.show()
```
